## Wat is hier al klaar?

- **Trino-client**: `from uwv_lab import trino; trino.query('SHOW CATALOGS')`
- **MinIO**: `from uwv_lab import s3; s3.ls('uwv-bronze')`
- **Hive Metastore-inspecties**: `from uwv_lab import hms; hms.list_tables('uwv')`
- **OpenMetadata API**: `from uwv_lab import om; om.tables(catalog='silver')`
- **Delta-rs**: rauwe Delta-tabellen lezen zonder Spark via
  `s3.read_delta('s3a://uwv-silver/wia/aanvraag')`.
- **Polars / DuckDB / Pandas** voor lokale analyse.
- **jupyterlab-git**: linker zijbalk → Git-icoon → commit & push.

## Wat moet je weten?

1. **Je identiteit telt.** Trino voert iedere query uit met jouw Keycloak
   `preferred_username`. OPA past row-filters, kolom-maskers en
   doelbinding toe op basis van je realm-rollen.
2. **Je werk wordt bewaard.** `~/work/` zit op een persistente volume,
   maar Git is de officiële versie-bron — commit vaak.
3. **Synthetische data.** Alle 10k cliënten zijn fictief. Geen echte
   BSN's, geen productie-koppelingen.
