# 04 — MinIO rauw doorzoeken

MinIO is de fysieke opslag onder het lakehouse. Voor verkenning en
debugging is `s3fs` handiger dan de MinIO-console: je kunt het direct
in een notebook script-en.

```python
from uwv_lab import s3

# Welke buckets bestaan?
s3.ls('')
```

```python
# Welke tabellen staan er in uwv-silver?
s3.ls('uwv-silver/')
```

```python
# Open een _delta_log JSON-entry — handig voor debug van partitionering.
fs = s3.fs()
for f in fs.ls('uwv-silver/uwv/persona_created/_delta_log/')[:3]:
    print(f)
    with fs.open(f, 'rb') as fh:
        print(fh.read()[:500])
    print('---')
```

```python
# Upload een klein bestand naar de staging-bucket.
fs = s3.fs()
with fs.open('uwv-staging/lab/scratch.txt', 'wb') as fh:
    fh.write(b'hello from uwv-lab\n')
print(fs.ls('uwv-staging/lab/'))
```
