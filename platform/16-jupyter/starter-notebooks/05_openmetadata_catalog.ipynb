# 05 — OpenMetadata vanuit een notebook

OpenMetadata is de catalog, glossary, lineage en data-quality hub van het
platform. Vanuit Python kun je dezelfde API bevragen die de OM-UI
gebruikt; nuttig om scripted DQ-checks, eigenaar-rapportages, of
lineage-traversals te bouwen.

```python
from uwv_lab import om

# Eerste 5 silver-tabellen — incl. owner + tags zoals OM die kent.
import pandas as pd
rows = om.tables(catalog='silver', limit=5)
pd.json_normalize(rows, max_level=1)[['name', 'fullyQualifiedName', 'tags', 'owner']]
```

```python
# Lineage opvragen — wat voedt deze tabel, en welke gold-tabellen lezen ervan?
lin = om.lineage_for('trino.silver.uwv.persona_created')
lin
```

```python
# Klikbare URL naar dezelfde tabel in de OM-UI.
om.ui_link('trino.silver.uwv.persona_created')
```

## Eigen glossary-term opzoeken

Glossaries en termen zitten in OM voor business-context. Met de API kun
je in een notebook een DQ-rapport schrijven dat een term tegen alle
kolommen mapt die ervan gebruik maken.

```python
for t in om.glossary_terms()[:10]:
    print(t['name'], '—', t.get('description', '')[:80])
```
