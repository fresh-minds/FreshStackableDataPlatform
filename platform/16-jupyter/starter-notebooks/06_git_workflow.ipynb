# 06 — Versie-beheer met Git

Notebooks in `~/work/` overleven pod-restart dankzij een PVC, maar **Git**
is hier de bron van waarheid. JupyterLab heeft een Git-extensie ingebouwd:
**linker zijbalk → Git-icoon** (de Octocat).

## Workflow (zonder de UI, vanuit een cel)

Werkflow voor een nieuw project:

```bash
cd ~/work
mkdir my-notebooks && cd my-notebooks
git init -b main
git remote add origin https://<your-pat>@github.com/<org>/<repo>.git
# kopieer een starter
cp ~/work/starter/01_trino_quickstart.ipynb ./
git add . && git commit -m 'first notebook'
git push -u origin main
```

Voor existing repo:

```bash
cd ~/work
git clone https://<your-pat>@github.com/<org>/<repo>.git
```

## PAT bewaren

We checken **geen** GitHub-credentials in. Twee veilige opties:

1. **Per-cel PAT in remote-URL** — zoals hierboven. Werkt voor één push,
   maar staat in `~/.git/config` op je PVC. Niet ideaal, wel pragmatisch.
2. **Git credential helper met store-met-eigenaar 0600** — eenmalig
   instellen, daarna onthoudt JupyterLab je credentials voor de duur van
   het pod. Je PAT staat in `~/.git-credentials`.

```bash
git config --global credential.helper store
# eerste push vraagt om username + PAT; daarna stilte.
```

## De UI

Voor dag-tot-dag werk is de Git-zijbalk waarschijnlijk fijner:

- **Onveranderd → Changes**: zie wat je hebt gewijzigd.
- **Right-click → Diff this notebook**: nbdiff (cell-level) i.p.v. JSON-diff.
- **Commit message + ✓**: maakt commit.
- **↑** rechtsboven: push.

Bekijken hoe je vorige versie eruitzag: **history**-tab in dezelfde
zijbalk.
